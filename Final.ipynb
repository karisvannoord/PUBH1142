The dataset utilized throughout this project reports the myriad factors that can contribute to the likelihood of an individual having a stroke. The dataset is from an unspecified source (Tutorial Data). Throughout an individuals life there are many variables that may predispose an individual to having a stroke. According to the National Institute of Health, "A stroke can occur when blood flow to the brain is blocked or there is sudden bleeding in the brain". 

There are two types of stroke that an individual can experience. A stroke that occurs because blood flow to the brain is blocked is called an ischemic stroke. The brain cannot get oxygen and nutrients from the blood. Without oxygen and nutrients, brain cells begin to die within minutes. A stroke that occurs because of sudden bleeding in the brain is called a hemorrhagic stroke. The leaked blood results in pressure on brain cells, damaging them. 

There are many variables that can lead to an individual experiencing a stroke within their lifetime. Some of the variables that can contribute to whether or not an indivial will experience a stroke are physical factors, such as gender, age, high blood pressure (hypertension), average glucose levels, and Body Mass Index (BMI). In addition to the physical factors that contribute to the prevalence of strokes, there are envrionmental factors that affect the likelihood of having a stroke such as smoking, marriage status, residence type, and employment status/type.

## Heart Disease

The next variable under consideration is heart disease. Heart disease can cause strokes through two mechanisms: 
- The formation of blood clots: many kinds of heart diease can lead to the formation of blood clots in the chambers of the heart. These clots can break loose and travel through the bloodstream where they can block blood flow to parts of the brain, causing strokes. This is called an ischemic stroke, and is the most common type of stroke. 
- Reduced blood pumping efficiency: heart disease can lead to reduced pumping efficiency of the heart. When the heart is unable to pump enough blood, the blood can pool in the heart chambers, increasing the risk of clots forming. Furthermore, reduced blood flow can lead to poor blood circulation, increasing the risk of the formation of clots throughout other parts of the body. 

- 275 of the participants had heart disease. 
- 4706 participants did not have heart disease.

## Average Glucose Levels 

The next variable to be considered is the average glucose levels of the participants. Consistently high levels of glucose in the blood, as seen in poorly controlled diabetes, contribute to the development of atherosclerosis, a condition where plaque builds up inside the arteries. Atherosclerosis narrows the arteries and can lead to the formation of blood clots, increasing the risk of ischemic strokes.
-  Chronic hyperglycemia (high blood sugar) is associated with the development of hypertension (high blood pressure). Hypertension is a major risk factor for strokes, as it damages blood vessels and increases the likelihood of blood clots forming. High glucose levels promote inflammation and endothelial dysfunction, which further contribute to the development of atherosclerosis and increase the risk of stroke.
-  While ischemic strokes are more common, elevated glucose levels can also contribute to hemorrhagic strokes, particularly in individuals with poorly controlled diabetes. High glucose levels can weaken blood vessel walls over time, making them more prone to rupture and causing bleeding in the brain. 

Overall, maintaining healthy glucose levels through lifestyle modifications, medication management (for individuals with diabetes), and regular monitoring is essential for reducing the risk of stroke and other cardiovascular complications. Controlling blood sugar levels can help prevent or delay the development of atherosclerosis, hypertension, and other conditions that increase the risk of stroke.

According to the Cleveland clinic, average glucose levels in a blood test should be between 70 and 99 mg/dL. Values between 50 and 70 mg/dL can also be considered normal among those who do not have diabetes. The measurement mg/dL refers to milligrams per deciliter. A deciliter measures fluid volume that is 1/10 of a liter.  
- Within the dataset, the lowest glucose level recorded was 55 mg/dL. 
- The highest glucose level recorded was 271 mg/dL.

## Body Mass Index

The next variable to be analyzed is body mass index. According to the CDC, Body Mass Index (BMI) is a persons weight in kilograms divided by the square of height in meters. A high BMI can indicate high body fatness. BMI screens for weight categories that may lead to health problems, but it does not diagnose the body fatness or health of an individual. 
- Obesity is strongly linked to hypertension, which is a leading risk factor for stroke. Excess body weight requires more blood to supply oxygen and nutrients to tissues, leading to increased blood volume and pressure on artery walls. Over time, this can damage blood vessels and increase the risk of stroke.
- Excess body fat, especially abdominal fat, can lead to insulin resistance, where cells become less responsive to insulin. This condition is a precursor to type 2 diabetes. Both insulin resistance and diabetes are associated with an increased risk of stroke due to their effects on blood vessel health, inflammation, and clot formation.
- Obesity often leads to a sedentary lifestyle, which is itself a risk factor for stroke. Lack of physical activity contributes to weight gain, hypertension, dyslipidemia, and other conditions that increase stroke risk.
